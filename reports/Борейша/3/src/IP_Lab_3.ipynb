```python
from ultralytics import YOLO

# model = YOLO('/content/drive/MyDrive/University/IP_Lab_3/blue_border/runs/yolov5n/weights/weights/last.pt')
model = YOLO('yolov5n.pt')

model.train(
    data='/content/IP_Lab_3/data.yaml',     # Путь к конфигурации датасета
    epochs=100,                       # Количество эпох
    imgsz=1024,                        # Размер изображения
    batch=16,                         # Размер батча
    save_period=5,
    verbose=True,
    plots=True,
    project='/content/drive/MyDrive/University/IP_Lab_3/blue_border/runs/yolov5n',  # Путь к директории для сохранения
    name='weights',                   # Название папки для весов
    resume=False                       # Продолжить обучение с последней точки
)
```
